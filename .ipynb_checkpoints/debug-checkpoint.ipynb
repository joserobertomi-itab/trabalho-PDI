# Segmentation pipeline debug

Runs the **full production pipeline** on a small sample (default: **one image per class**):

1. `load_image` → `preprocess_image` → masks → candidates → score → postprocess → `crop_and_save`
2. Writes graded-style crops to `debug_result/result/`
3. Writes per-frame diagnostics to `debug_result/bundles/`

CLI equivalent: `make debug` or `uv run pdiseg-debug`.

Do **not** write debug artifacts into `result/` (that folder is for delivery).

## 1. Setup

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import pdiseg

ROOT = Path('.').resolve()
DATASET = ROOT / 'data' / 'Train_and_Validation'
DEBUG_RESULT = ROOT / 'debug_result'
PER_CLASS = 1  # increase to 2–3 for more coverage

if not DATASET.is_dir():
    raise FileNotFoundError(f'Dataset not found: {DATASET}')

print(f'Classes: {len(pdiseg.list_classes(DATASET))}')
print(f'Sample size: up to {PER_CLASS} image(s) per class')
DATASET
```

## 2. Run full pipeline on sample

Uses the same detector and crop writer as `make run`, but only on the selected sample.

```python
sample_report = pdiseg.run_debug_sample(
    DATASET,
    DEBUG_RESULT / 'result',
    bundle_root=DEBUG_RESULT / 'bundles',
    per_class=PER_CLASS,
)

ds = sample_report.dataset_report
print(f'frames={ds.total_frames} crops={ds.total_crops} empty={ds.empty_frames}')
print(f'avg crops/frame={ds.avg_crops_per_frame:.2f}')
print(f'result -> {sample_report.result_root}')
print(f'bundles -> {sample_report.bundle_root}')

summary = pd.DataFrame(
    [
        {
            'class': row.class_name,
            'frames': row.frames,
            'crops': row.crops,
            'empty': row.empty_frames,
        }
        for row in ds.classes
    ]
)
summary
```

## 3. Inspect one frame (all stages)

Pick the first processed frame and walk through preprocess → masks → detection stages.

```python
if not sample_report.frames:
    raise RuntimeError('No frames processed — check dataset path and PER_CLASS')

frame = sample_report.frames[0]
image = pdiseg.load_image(frame.source)
detection, prep, masks = pdiseg.debug_frame(image)

print('class:', frame.class_name)
print('file:', frame.source.name)
print('candidates:', len(detection.candidates))
print('kept:', len(detection.kept))
print('labels:', len(detection.labels))
print('saved crops:', len(frame.labels))

scored_df = pd.DataFrame(pdiseg.scored_table(detection.scored))
display_cols = [
    c
    for c in ['rank', 'score', 'box', 'dark_density', 'bright_on_dark', 'background_level', 'extent']
    if c in scored_df.columns
]
scored_df[display_cols].head(12)
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes[0, 0].imshow(pdiseg.to_rgb(image))
axes[0, 0].set_title('original')
axes[0, 1].imshow(prep.work, cmap='gray')
axes[0, 1].set_title('preprocessed (CLAHE + FPS mask)')
axes[0, 2].imshow(pdiseg.draw_boxes(image, detection.candidates, (220, 50, 50)))
axes[0, 2].set_title(f'raw candidates ({len(detection.candidates)})')

mask_views = pdiseg.visualize_masks(masks)
axes[1, 0].imshow(mask_views['combined'], cmap='gray')
axes[1, 0].set_title('combined mask')
axes[1, 1].imshow(pdiseg.draw_boxes(image, detection.kept, (240, 200, 40)))
axes[1, 1].set_title(f'kept after score ({len(detection.kept)})')
axes[1, 2].imshow(pdiseg.draw_boxes(image, detection.labels, (40, 200, 80)))
axes[1, 2].set_title(f'final labels ({len(detection.labels)})')

for ax in axes.ravel():
    ax.axis('off')
plt.tight_layout()
```

```python
if detection.labels:
    ncols = len(detection.labels)
    fig, axes = plt.subplots(1, ncols, figsize=(3 * ncols, 3))
    axes = [axes] if ncols == 1 else list(axes)
    for i, box in enumerate(detection.labels, start=1):
        axes[i - 1].imshow(pdiseg.crop(image, box), cmap='gray')
        axes[i - 1].set_title(f'crop {i}')
        axes[i - 1].axis('off')
    plt.tight_layout()
else:
    print('no labels for this frame')
```

## 4. Top scorer feature summary

```python
for line in pdiseg.feature_summary(detection.scored):
    print(line)
```

## 5. Browse sample across classes

Each row shows production crop count and the on-disk debug bundle folder.

```python
rows = []
for item in sample_report.frames:
    rows.append(
        {
            'class': item.class_name,
            'file': item.source.name,
            'labels': len(item.labels),
            'bundle': str(item.bundle_dir),
        }
    )
pd.DataFrame(rows)
```

## 6. Optional: inspect a specific class

Change `TARGET_CLASS` to drill into one product folder.

```python
TARGET_CLASS = sample_report.frames[0].class_name

matches = [f for f in sample_report.frames if f.class_name == TARGET_CLASS]
if not matches:
    print(f'No sample frame for class {TARGET_CLASS!r}')
else:
    item = matches[0]
    img = pdiseg.load_image(item.source)
    det, prep, masks = pdiseg.debug_frame(img)
    overlay = pdiseg.render_overlay(
        img,
        pdiseg.FrameInspection(
            candidates=det.candidates,
            kept=det.kept,
            labels=det.labels,
        ),
    )
    plt.figure(figsize=(10, 6))
    plt.imshow(overlay)
    plt.title(f'{TARGET_CLASS} — red=rejected yellow=kept green=final')
    plt.axis('off')
```
